First, note the rate of change of the option price is the slope of the tangent line to the curves displayed in the figures below. 

In addition, we note that the time sensitivity of an option is known as the **theta** of an option,

$$\Theta_{C_0} \coloneqq \frac{\partial C_0}{\partial t}\quad \mbox{and}\quad \Theta_{P_0} \coloneqq \frac{\partial P_0}{\partial t},$$

while the spot price sensitivity is the **delta** of an option,

$$\Delta_{C_0}:=\frac{\partial C_0}{\partial S_0} \quad \mbox{and}\quad \Delta_{P_0}:=\frac{\partial C_0}{\partial S_0}.$$

Thus, we are to comment on the delta and theta of call and put options.

### Code:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
import seaborn as sns
sns.set_style('darkgrid')

def bs_call(S0, K, sigma, t=1, r=0):
    """
    Description:
    
    Computes the Black-Scholes value of a European call option.
    
    Parameters:
        S0: Current asset price
        K: Strike price
        sigma: Yearly standard deviation of log-returns (volatility)
        t: Time to expiration (in years)
        r: Risk-free interest rate
    
    Returns:
        Call option price
    """
    d1 = (np.log(S0/K) + (r+.5*sigma**2)*t)/(sigma*np.sqrt(t))
    #d2 = (np.log(S0/K) + (r-.5*sigma**2)*t)/(sigma*np.sqrt(t))
    d2 = d1 - sigma*np.sqrt(t)
    return S0*norm.cdf(d1)-K*np.exp(-r*t)*norm.cdf(d2)


def bs_put(S0, K, sigma, t=1, r=0):
    """
    Description:
    
    Computes the Black-Scholes value of a European put option.
    
    Parameters:
        S0: Current asset price
        K: Strike price
        sigma: Yearly standard deviation of log-returns (volatility)
        t: Time to expiration (in years)
        r: Risk-free interest rate
    
    Returns:
        Put option price
    """
    
    d1 = (np.log(S0/K) + (r+.5*sigma**2)*t)/(sigma*np.sqrt(t))
    #d2 = (np.log(S0/K) + (r-.5*sigma**2)*t)/(sigma*np.sqrt(t))
    d2 = d1 - sigma*np.sqrt(t)
    return -S0*norm.cdf(-d1)+K*np.exp(-r*t)*norm.cdf(-d2)
```

```python
# Plot call price as a function of time to expiration t (with underlying spot price fixed at S0=100) 
# and as a function of underlying spot price S0 (with underlying time to expiration fixed at t=1)

# In what follows, for consistency we will always take the following values of sigma and K:

K = 110
sigma = .3

# As a function of time to expiration t

S0 = 100
times = np.linspace(1/12, 40, 200)

call_prices = [bs_call(S0,K,sigma, t = times[i]) for i in range(len(times))]
put_prices = [bs_put(S0,K,sigma, t = times[i]) for i in range(len(times))]


fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].plot(times, call_prices, label="Call price")
axes[0].plot(times, put_prices, label="Put price")
axes[0].set_title('Black-Scholes Options prices vs time to expiration', size = 16)
axes[0].set_xlabel('Time to Expiration')
axes[0].set_ylabel('Black-Scholes Options Price')
axes[0].legend()

# As a function of S0

S0 = np.linspace(50, 200, 221)
time = 1

call_prices1 = [bs_call(S0[i],K,sigma, t = time) for i in range(len(S0))]
put_prices1 = [bs_put(S0[i],K,sigma, t = time) for i in range(len(S0))]


axes[1].plot(S0, call_prices1, label="Call price")
axes[1].plot(S0, put_prices1, label="Put price")
axes[1].set_title('Black-Scholes Options prices vs spot price', size = 16)
axes[1].set_xlabel('Spot Price')
axes[1].set_ylabel('Black-Scholes Options Price')
axes[1].legend()

plt.show()
```

### Observations:

We suspect that the call and put thetas (the slopes of the curves on the left plot) are monotone decreasing functions of the time to expiration which satisfy
$$ \lim_{t\to 0^+}\Theta_{C_0}(t) = \lim_{t\to 0^+}\Theta_{P_0}(t) = -\infty, \qquad \lim_{t\to +\infty}\Theta_{C_0}(t) = \lim_{t\to +\infty}\Theta_{P_0}(t) = c \geq 0$$
for some non-negative real number $c$.

On the other hand, for the call and put deltas (the slopes of the curves on the right plot), we suspect that both the call and put deltas are monotone increasing functions of $S_0$ which satisfy

$$\lim_{S_0\to 0^+}\Delta_{C_0}(S_0) = 0, \quad \lim_{S_0\to +\infty} \Delta_{C_0}(S_0) = 1, \quad \lim_{S_0\to 0^+} \Delta_{P_0}(S_0) = -1, \quad \lim_{S_0\to +\infty}\Delta_{P_0}(S_0) = 0.$$

### Comparison with analytical formulas:

We know from Homework 3 that:

$$\Delta_{C_0} = \Phi(d_1), \qquad \Delta_{P_0}= \Phi(d_1)-1,$$

and

$$\begin{align}
\Theta_{C_0} &= - \frac{S_0\varphi(d_1)\sigma}{2\sqrt{t}} - rKe^{-rt}\Phi(d_2), \\
\Theta_{P_0} &= - \frac{S_0\varphi(d_1)\sigma}{2\sqrt{t}} + rKe^{-rt}\Phi(-d_2).
\end{align}$$

These formulas confirm the qualitative behaviour described in our observations, as the following plots show.

```python
# Define the deltas of call and put options


def bs_call_delta(S0, K, sigma, t, r):
    """
    Returns the Delta (sensitivity to spot price) of a European call option
    under Black-Scholes assumptions.

    Parameters:
        S0 (float): Initial stock price
        K (float): Strike price
        sigma (float): Volatility of the stock
        t (float): Time to maturity (in years)
        r (float): Risk-free interest rate

    Returns:
        float: Delta of Call Option
    """
    d1 = (np.log(S0/K) + (r+.5*sigma**2)*t)/(sigma*np.sqrt(t))
    return norm.cdf(d1)


def bs_put_delta(S0, K, sigma, t, r):
    """
    Returns the Delta (sensitivity to spot price) of a European put option
    under Black-Scholes assumptions.

    Parameters:
        S0 (float): Initial stock price
        K (float): Strike price
        sigma (float): Volatility of the stock
        t (float): Time to maturity (in years)
        r (float): Risk-free interest rate

    Returns:
        float: Delta of Put Option
    """
    d1 = (np.log(S0/K) + (r+.5*sigma**2)*t)/(sigma*np.sqrt(t))
    return norm.cdf(d1) - 1.



# Define the thetas of call and put options


def bs_call_theta(S0,K,sigma,t,r):
    """
    Returns the theta (rate of change with respect to time) of a European call option
    under Black-Scholes assumptions.
    
    
    To give daily theta, divide return by 252.

    Parameters:
        S0 (float): Initial stock price
        K (float): Strike price
        sigma (float): Volatility of the stock
        t (float): Time to maturity (in years)
        r (float): Risk-free interest rate

    Returns:
        float: Theta of Call option
    """
    
    d1 = (np.log(S0/K) + (r+.5*sigma**2)*t)/(sigma*np.sqrt(t))
    d2 = d1 - sigma*np.sqrt(t)
    
    return (-S0*norm.pdf(d1)*sigma)/(2*np.sqrt(t)) - r*K*np.exp(-r*t)*norm.cdf(d2)


def bs_put_theta(S0,K,sigma,t,r):
    """
    Returns the theta (rate of change with respect to time) of a European put option
    under Black-Scholes assumptions.
    
    
    To give daily theta, divide return by 252.

    Parameters:
        S0 (float): Initial stock price
        K (float): Strike price
        sigma (float): Volatility of the stock
        t (float): Time to maturity (in years)
        r (float): Risk-free interest rate

    Returns:
        float: Theta of Put option
    """
    
    d1 = (np.log(S0/K) + (r+.5*sigma**2)*t)/(sigma*np.sqrt(t))
    d2 = d1 - sigma*np.sqrt(t)
    
    return (-S0*norm.pdf(d1)*sigma)/(2*np.sqrt(t)) + r*K*np.exp(-r*t)*norm.cdf(-d2)


# Thetas as a function of time to expiration t

S0 = 100
times = np.linspace(1/12, 40, 200)

call_theta = [bs_call_theta(S0,K,sigma, t = times[i],r=0) for i in range(len(times))]
put_theta = [bs_put_theta(S0,K,sigma, t = times[i],r=0) for i in range(len(times))]

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].plot(times, call_theta, label="Call theta")
axes[0].plot(times, put_theta, label="Put theta")
axes[0].set_title('Theta of B-S Options prices vs time to expiration', size = 16)
axes[0].set_xlabel('Time to Expiration')
axes[0].set_ylabel('Theta of B-S Options Price')
axes[0].legend()

# Deltas as a function of S0

S0 = np.linspace(50, 200, 221)
time = 1

call_delta = [bs_call_delta(S0[i],K,sigma, t = time,r=0) for i in range(len(S0))]
put_delta = [bs_put_delta(S0[i],K,sigma, t = time,r=0) for i in range(len(S0))]

axes[1].plot(S0, call_delta, label="Call delta")
axes[1].plot(S0, put_delta, label="Put delta")
axes[1].set_title('Delta of B-S Options prices vs spot price', size = 16)
axes[1].set_xlabel('Spot Price')
axes[1].set_ylabel('Delta of B-S Options Price')
axes[1].legend()

plt.show()
```
